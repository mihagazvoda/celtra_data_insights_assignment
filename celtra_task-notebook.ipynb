## 2. Calculate the overall effective Ad engagement rate, that is the percentage of sessions that contain interaction or firstInteraction event.
Ad engagement rate was calculated by counting unique sessions which contained interaction or firstInteraction and then dividing them with the number of all sessions. 
Calculating confidence intervals I made next assumptions: 
* The experiment consists of identical trials.
* Each trial results in one of the two outcomes, called success (probability p) and failure (probability q=1-p).
* The probability of success, denoted p remains the same from trial to trial.
* The n trials are independent.
* If the sample is large enough  Binomial distribution can be approximated by  normal if the sample is large enough (np>5 & nq>5).

```python
engaged = df.loc[(df['name'] == 'interaction') | (df['name'] == 'firstInteraction'), 'sessionId'].nunique()
seen = df['sessionId'].nunique()

er = engaged / seen
std_er = np.sqrt(er * (1 - er) / seen)
bound = 1.96*100*std_er


print(f"Engagement rate is {format(100 * er, '.2f')}% " 
      f"with the 95% confidence interval [{format(er*100-bound, '.2f')}%, {format(bound+er*100, '.2f')}%].")
```

## 4.1 Can you find that point and check whether the difference is statistically significant?


I created data frame (df_engagement) which consists of sessionIds, timestamps of adRequested events and engagement booleans.  
I searched for the border time with the biggest (most statistically significant) change using for loop with 10-minute interval. Procedure:
* split df_engagement on the selected border time
* calculate mean and standard deviation of ER before and after border time
* binomial distribution turns into normal for large samples (described in the previous paragraph)
* use z-test with the null hypothesis that sample means are equal and calculate p-value

I plotted results on the graph below. Biggest difference coincide with the smallest p-value so that's the border time (p-values are more important than ER differences).   
In case I want better precision I could make a for loop with shorter interval. For even better precision I could fit the function on the p-values around the minimum and then search for it's minimum analytically or by using scipy.optimize.minimize.

```python
df_engagement = engagement_df(df)
df_engagement.head()
```

```python
# linearly spaced data for searching the time with the largest (most significant) ER diference
date_range = pd.date_range(start, stop, freq='10min')[1:-1]

# create empty df
er_df = pd.DataFrame(index=date_range, columns=['diff','p_value', 'k1', 'k2'], dtype='float64')

for border_time in date_range:
    er_df.loc[border_time] = er_stats(df_engagement, border_time)

er_df.head()
```

```python
fig,ax = plt.subplots()

ax.plot((er_df['diff']*100).index, (er_df['diff']*100).values)
ax.set(xlabel='border time', ylabel='ER difference in %')
ax.legend(['ER difference'], loc=2, frameon=True)

xfmt = md.DateFormatter('%H:%M')
ax.xaxis.set_major_formatter(xfmt)

ax2 = ax.twinx()
ax2.plot(er_df['p_value'], color='black', alpha=0.5)
ax2.set(yscale="log", ylabel='P-value', title='ER differences and P-values for different border times')

ax2.legend(['P-value'], loc=1, frameon=True);
ax2.grid(False);
ax2.xaxis.set_major_formatter(xfmt)
```

```python
max_diff_time = er_df['p_value'].idxmin()
print(f'Time with the smallest p-value: {max_diff_time}')
```
